```python
def gradient_descent(x,y,learning_rate=0.01,epochs=100):
  cost_hist=[]
  m_curr=b_curr=0
  iteration=1000
  n=len(x)
  learning_rate=0.08
  for i in range(iteration):
    yp=m_curr*x+b_curr
    cost=(1/n)*sum([val**2 for val in (y-yp)])
    md=-(2/n)*sum(x*(y-yp))
    bd=-(2/n)*sum(y-yp)
    b_curr=b_curr-learning_rate*bd
    m_curr=m_curr-learning_rate*md
    cost_hist.append(cost)
    print(f"m:{m_curr},b:{b_curr},cost:{cost},iteration:{i}")
  return cost_hist

cost_hist=gradient_descent(x,y)
```

```python
# Given small dataset
x = np.array([[1], [2], [3], [4], [5]])  # Input feature
y = np.array([[5], [7], [9], [11], [13]])  # Actual values (y = 2x + 3, no noise)
```

```python
x
```

```python
y
```

```python
# Add bias term (column of ones)
X_b = np.c_[np.ones((5, 1)), x] #np.ones((5, 1))This creates a column vector of ones with 5 rows and 1 column.It serves as the bias term (intercept)
                                # in a linear regression model.
                                #np.c_[...] (Concatenation) np.c_[] is a shortcut for column-wise concatenation.
                                #It appends the column of ones to x, forming an augmented matrix.The result is a new matrix with x values and a bias term.
```

```python
X_b
```

```python
# Define the analytical cost function
def analytical_cost(theta, x, y):
    m = len(y)
    predictions = x.dot(theta)
    cost = (1 / (2 * m)) * np.sum(np.square(predictions - y))
    return cost
```

```python
# Compute cost for different theta values
theta_0_values = np.linspace(0, 6, 50)  # Try different intercepts
theta_1_values = np.linspace(0, 4, 50)  # Try different slopes
```

```python
# Compute cost for each (theta_0, theta_1) pair
Theta_0, Theta_1 = np.meshgrid(theta_0_values, theta_1_values)
Cost = np.zeros(Theta_0.shape)
# np.meshgrid(theta_0_values, theta_1_values)
# Purpose:
# Creates a grid of values for theta_0 and theta_1, which are parameters in a linear regression model.
# This is useful for visualizing the cost function across different parameter values.
```

```python
for i in range(len(theta_0_values)):
    for j in range(len(theta_1_values)):
        theta_test = np.array([[theta_0_values[i]], [theta_1_values[j]]])
        Cost[j, i] = analytical_cost(theta_test, X_b, y)
```

```python
# Plot the cost function as a heatmap
plt.figure(figsize=(8, 6))
plt.contourf(Theta_0, Theta_1, Cost, levels=50, cmap='coolwarm')
plt.colorbar(label="Cost J()")
plt.xlabel("Theta_0 (Intercept)")
plt.ylabel("Theta_1 (Slope)")
plt.title("Cost Function Contour")

# Find optimal theta using normal equation
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
optimal_cost = analytical_cost(theta_best, X_b, y)

# Mark the optimal theta
plt.scatter(theta_best[0, 0], theta_best[1, 0], color='red', marker='x', s=100, label="Optimal _best")
plt.legend()
plt.show()

print("Optimal Theta (Intercept, Slope):", theta_best.ravel())
print("Minimum Cost J(_best):", optimal_cost)
```

```python
#If you want to minimize()using gradient descent, you would do something like:
alpha = 0.01  # Learning rate
theta = np.random.randn(2, 1)  # Initialize randomly
m = len(y)

# Gradient Descent
for _ in range(1000):
    gradients = (1/m) * X_b.T.dot(X_b.dot(theta) - y)
    theta -= alpha * gradients

print("Theta found by Gradient Descent:", theta.ravel())
print("Minimum Cost J(_best):", analytical_cost(theta, X_b, y))
```

```python
def stochastic_gradient_descent(x,y,theta,learning_rate=0.01,iterations=100):
  m=len(x)
  cost_history=np.zeros(iterations*m)
  theta_history=np.zeros((iterations*m,theta.shape[0]))
  for it in range(iterations):
        for i in range(m):
            rand_index=np.random.randint(0,m)
            x_i=x[rand_index, : ].reshape(1,x.shape[1])
            y_i=y[rand_index, : ].reshape(1,1)
            prediction=x_i.dot(theta)
            theta=theta-learning_rate*(x_i.T.dot(prediction-y_i))
             ### Updating theta and cost values arrays
            cost=analytical_cost(theta,x,y)
            theta_history[it*m+i, :]=theta.T
            #theta_history[it*m + i, :] = theta.T
            cost_history[it*m+i]=cost
  return theta, cost_history, theta_history
```

```python
#initialize theta
theta=np.random.randn(2,1)
x_b=np.c_[np.ones((len(x),1)),x]
# Gradient descent parameters
learning_rate = 0.01
iterations =1000
```

```python
#call the function
theta_sgd, cost_history_sgd, theta_history_sgd = stochastic_gradient_descent(x_b, y, theta,  learning_rate, iterations)
```

```python
print('SGD Theta0:          {:0.3f},\nSGD Theta1:          {:0.3f}'.format(theta_sgd[0][0], theta_sgd[1][0]))
print('SGD Final cost/MSE:  {:0.3f}'.format(cost_history_sgd[-1]))
```

batch gradient descent by j

```python
def batch_gradient_descent(x,y,theta,learning_rate=0.01,iterations=100):
  m=len(y)
  cost_history=np.zeros(iterations)
  theta_history=np.zeros((iterations,theta.shape[0]))
  for it in range(iterations):
    prediction=x.dot(theta)
    theta=theta-(1/m)*learning_rate*(x.T.dot(prediction-y))
    #updating arrays with theta/cost values
    theta_history[it, :]=theta.T

    cost_history[it]=analytical_cost(theta,x,y)
  return theta, cost_history, theta_history
```

```python
#call
theta_bgd, cost_history_bgd, theta_history_bgd = batch_gradient_descent(x_b, y, theta, learning_rate, iterations)
```

```python
print('bGD Theta0:          {:0.3f},\nbGD Theta1:          {:0.3f}'.format(theta_bgd[0][0], theta_bgd[1][0]))
print('bGD Final cost/MSE:  {:0.3f}'.format(cost_history_bgd[-1]))
```

mini batch GD

```python
def mini_batch_gradient_descent(x,y,theta,learning_rate=0.01,iterations=100,batch_size=32):
  m=len(y)
  cost_history=np.zeros(iterations)
  theta_history=np.zeros((iterations,theta.shape[0]))
  n_batches = m // batch_size
  for it in range (iterations):

    for i in range(0,m,batch_size):
      x_i=x[i:i+batch_size]
      y_i=y[i:i+batch_size]
      prediction=x_i.dot(theta)
      theta=theta-(1/batch_size)*learning_rate*(x_i.T.dot(prediction-y_i))
      theta_history[it, :]=theta.T
      cost_history[it]=analytical_cost(theta,x,y)
  return theta, cost_history, theta_history
```
