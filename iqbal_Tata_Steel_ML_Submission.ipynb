# **Project Summary -**

Introduction
The steel manufacturing industry faces significant challenges due to unexpected machine failures, leading to considerable downtime and financial losses. Tata Steel, a leader in the industry, embarked on a project to leverage predictive maintenance through machine learning to enhance operational efficiency and reduce costs. This project aimed to develop a robust machine learning model capable of accurately predicting machine failures, enabling timely maintenance and minimizing unplanned downtimes.

Data Collection and Preprocessing
The project began with the collection of extensive historical data on machine operations, failures, and maintenance records from Tata Steel's manufacturing units. Additional relevant data, such as environmental conditions, machine specifications, and usage patterns, were also gathered to provide a comprehensive dataset for analysis.

Data preprocessing was a critical step, involving cleaning and preparing the data to handle missing values, outliers, and noise. Numerical features were normalized to ensure they were on a similar scale, and interaction features were created to capture potential relationships between variables. This step ensured that the data was suitable for training machine learning models.

Feature Selection and Handling Imbalanced Data
To improve model performance and reduce the risk of overfitting, feature selection techniques were employed. A correlation matrix and hypothesis testing were used to identify the most relevant features. This process helped in selecting variables that had a significant impact on predicting machine failures.

Given the imbalanced nature of the dataset, with fewer instances of machine failures compared to normal operations, techniques such as Synthetic Minority Over-sampling Technique (SMOTE) were applied. SMOTE helped in balancing the dataset by generating synthetic samples for the minority class, improving the model's ability to detect machine failures.

Model Implementation and Evaluation
Several machine learning models were implemented and evaluated for their performance. Models such as Random Forest, XGBoost, and LightGBM were chosen due to their robustness and effectiveness in handling complex datasets. Each model's performance was evaluated using metrics such as precision, recall, F1-score, and accuracy to ensure a comprehensive assessment.

Hyperparameter tuning was conducted using GridSearchCV and RandomizedSearchCV to optimize the models' parameters and enhance their performance. Cross-validation techniques were employed to validate the models, ensuring their robustness and reliability.

Pilot Implementation and Results
The most effective model was selected for a pilot implementation within Tata Steel's manufacturing units. The model's predictions were monitored and compared against actual machine performance. The pilot implementation demonstrated that the model could accurately predict machine failures, allowing for timely maintenance and reducing unexpected downtimes.

The results showed a significant reduction in maintenance costs and increased operational efficiency. The model's predictions enabled Tata Steel to optimize the allocation of maintenance resources, ensuring that critical machines received attention when needed. This proactive approach to maintenance also enhanced workplace safety by ensuring machines were in good working condition, protecting both employees and equipment.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python

train_df.info()
````

````python
train_df.head()
````

````python
#convert categorical to numeric
train_df = pd.get_dummies(train_df, columns=['Type'], drop_first=True)
test_df=pd.get_dummies(test_df,columns=['Type'],drop_first=True)
````

````python
# Normalize numerical features
numerical_features = ['Air temperature [K]', 'Process temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]']
train_df[numerical_features] = (train_df[numerical_features] - train_df[numerical_features].mean()) / train_df[numerical_features].std()
test_df[numerical_features] = (test_df[numerical_features] - test_df[numerical_features].mean()) / test_df[numerical_features].std()
````

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Hypothesis:** There is no significant difference in process temperature between failed and non-failed machines.
**Null Hypothesis (H0):** The mean process temperature for failed machines is equal to the mean process temperature for non-failed machines.

Accuracy: Indicates the overall correctness of the model. Higher accuracy means fewer misclassifications, leading to better reliability in predicting machine failures.

Precision: Indicates the proportion of correctly predicted positive observations to the total predicted positives. High precision reduces false alarms, saving maintenance costs.

ROC-AUC: Measures the model's ability to distinguish between classes. A high AUC indicates better performance in identifying machine failures.

Business Impact: Accurate predictions enable timely maintenance, optimal resource allocation, cost savings, improved safety, and higher productivity.
